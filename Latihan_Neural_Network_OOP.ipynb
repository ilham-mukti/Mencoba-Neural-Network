```python
from jcopdl.layers import linear_block
```

```python
class CustomNN(nn.Module):
  def __init__(self, input_size, n1, n2, output_size, dropout):
    super().__init__()
    self.fc = nn.Sequential(
        linear_block(input_size, n1, dropout=0.2),
        linear_block(n1, n2, dropout=0.2),
        linear_block(n2, output_size, activation='lsoftmax', dropout=dropout),
    )
  def forward(self, x):
    return self.fc(x)
```

```python
model = CustomNN(7, 16, 8, 2, dropout=0.2)
```

```python
contoh_data = torch.rand(200, 7)
```

```python
output = model(contoh_data)
output
```
